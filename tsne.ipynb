```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd # para leer datos
import sklearn.manifold
import sklearn.datasets
import glob
import sklearn.cluster

```

Se incluye un Dataset que clasifica hojas de plantas de acuerdo a su perfil

```python
data = pd.read_csv('data_Mar_64.txt')
columns_num= np.arange(0,65)

data.columns = columns_num

vector = data.drop([0],axis=1)
planta = data[0]

X = vector


planta_str = data[0].str.split().str[0]

Y = planta_str

n_plantas = len(planta_str)

y_numbers = pd.factorize(Y)
```

```python
for i in range(2):
    tsne = sklearn.manifold.TSNE(perplexity=(i*20+5))
    tsne.fit(X)
    embedding = tsne.embedding_
    
    plt.figure(figsize=(20,5))
    plt.subplot(1,2,i+1)
    plt.scatter(embedding[:,0], embedding[:,1], c=y_numbers[0], cmap='rainbow', s=30.0)
    plt.colorbar(boundaries=np.arange(34)-0.5).set_ticklabels(y_numbers[1])
```

Clasifica las hojas aceptablemente dentro de clusters pero dentro de una clase de hojas hay subgrupos
